```python
#Task 4

import matplotlib.pyplot as plt
import numpy as np

temperatures = [8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
                16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
                13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
                7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
                16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5]


night_temps = []   
evening_temps = [] 
day_temps = []     

for i in range(0, len(temperatures), 3):
    night_temps.append(temperatures[i])
    evening_temps.append(temperatures[i+1])
    day_temps.append(temperatures[i+2])

avg_night_temp = round(np.mean(night_temps), 2)
avg_evening_temp = round(np.mean(evening_temps), 2)
avg_day_temp = round(np.mean(day_temps), 2)

print("Temperature Analysis by Time of Day:")
print(f"Night Temperatures (00-08): {night_temps}")
print(f"Average Night Temperature: {avg_night_temp} ")
print(f"\nEvening Temperatures (08-16): {evening_temps}")
print(f"Average Evening Temperature: {avg_evening_temp}")
print(f"\nDay Temperatures (16-24): {day_temps}")
print(f"Average Day Temperature: {avg_day_temp}")

plt.figure(figsize=(10, 6))
days = list(range(1, len(day_temps) + 1))
plt.scatter(days, day_temps, color='red', label='Day Temperatures')
plt.plot(days, day_temps, color='lightcoral', linestyle='--')
plt.title('Day Temperatures at TIA, Kathmandu')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#Task 5

def sum_nested_list(nested_list):
    total = 0

    for item in nested_list:
        if isinstance(item, list):
            total += sum_nested_list(item)
        elif isinstance(item, (int, float)):
            total += item

    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]

print("Nested List:", nested_list)
print("Total Sum:", sum_nested_list(nested_list))
```

```python
#Task 6

def generate_permutations(s):

    chars = list(s)

    result = []

    def backtrack(start):
        if start == len(chars):
            result.append(''.join(chars))
            return

        used = set()

        for i in range(start, len(chars)):
            if chars[i] in used:
                continue

            used.add(chars[i])

            chars[start], chars[i] = chars[i], chars[start]

            backtrack(start + 1)

            chars[start], chars[i] = chars[i], chars[start]

    backtrack(0)
    return result

print("Permutations of 'abc':")
print(generate_permutations("abc"))

print("\nPermutations of 'aab':")
print(generate_permutations("aab"))
```

```python
#Task 7

def calculate_directory_size(directory):

    total_size = 0

    for item, value in directory.items():
        if isinstance(value, int):
            total_size += value

        elif isinstance(value, dict):
            total_size += calculate_directory_size(value)

        else:
            print(f"Warning: Unexpected item type for {item}")

    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print("Total directory size:")
total_size = calculate_directory_size(directory_structure)
print(f"{total_size} KB")

print("\nTest Cases:")
print("Empty directory:", calculate_directory_size({}))

single_level_dir = {
    "doc1.txt": 500,
    "doc2.txt": 750
}
print("Single-level directory:", calculate_directory_size(single_level_dir))

deeply_nested_dir = {
    "level1": {
        "level2": {
            "level3": {
                "file.txt": 1000
            },
            "another_file.txt": 200
        }
    },
    "top_level_file.txt": 300
}
print("Deeply nested directory:", calculate_directory_size(deeply_nested_dir))
```

```python
#Task 8

def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)

    dp[0] = 0

    for current_amount in range(1, amount + 1):
        for coin in coins:
            if coin <= current_amount:
                dp[current_amount] = min(
                    dp[current_amount],
                    dp[current_amount - coin] + 1
                )

    return dp[amount] if dp[amount] != float('inf') else -1

# Test cases
def test_min_coins():
    coins = [1, 2, 5]
    amount = 11
    result = min_coins(coins, amount)
    print(f"Coins {coins}, Amount {amount}:")
    print(f"Minimum coins needed: {result}")
    print("Expected: 3 (5 + 5 + 1)")

    coins = [2]
    amount = 3
    result = min_coins(coins, amount)
    print(f"\nCoins {coins}, Amount {amount}:")
    print(f"Minimum coins needed: {result}")
    print("Expected: -1 (impossible)")

    coins = [186, 419, 83, 408]
    amount = 6249
    result = min_coins(coins, amount)
    print(f"\nCoins {coins}, Amount {amount}:")
    print(f"Minimum coins needed: {result}")

# Running test cases
test_min_coins()
```

```python
#Task 9

def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1

            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

def reconstruct_lcs(s1, s2):
    m, n = len(s1), len(s2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])


    lcs = []
    i, j = m, n
    while i > 0 and j > 0:
        if s1[i-1] == s2[j-1]:
            lcs.append(s1[i-1])
            i -= 1
            j -= 1
        elif dp[i-1][j] > dp[i][j-1]:
            i -= 1
        else:
            j -= 1

    return ''.join(reversed(lcs))

def test_lcs():
    test_cases = [
        ("abcde", "ace"),
        ("abc", "abc"),
        ("abc", "def"),
        ("AGGTAB", "GXTXAYB"),
        ("", "abc"),
        ("", "")
    ]

    for s1, s2 in test_cases:
        print(f"\nStrings: '{s1}' and '{s2}'")

        lcs_length = longest_common_subsequence(s1, s2)
        print(f"LCS Length: {lcs_length}")

        if lcs_length > 0:
            lcs = reconstruct_lcs(s1, s2)
            print(f"LCS: '{lcs}'")
        else:
            print("No LCS found.")

test_lcs()
```

```python
#Task 10

def knapsack(weights, values, capacity):
    n = len(weights)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            current_weight = weights[i-1]
            current_value = values[i-1]

            if current_weight > w:
                dp[i][w] = dp[i-1][w]
            else:
                dp[i][w] = max(
                    dp[i-1][w],
                    dp[i-1][w - current_weight] + current_value
                )

    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

result = knapsack(weights, values, capacity)
print(f"Maximum value: {result}")
```
