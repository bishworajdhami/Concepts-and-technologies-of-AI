```python
#3. For the column with missing values fill the values using various techniques we discussed above. Try to explain why did you select the particular methods for particular column.


import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Concepts and Technology of AI/medical_students_dataset.csv')

# Filling Missing Values Using Different Techniques
# Display the first few rows to understand the data
print("\nPreview of Dataset:")
print(df.head())

# Example decisions for handling missing values:
# - Numerical columns: Fill with mean or median
# - Categorical columns: Fill with mode

# Get a list of numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Fill missing values for numerical columns with mean
for col in numerical_cols:
    if df[col].isnull().any():  # Check if column has missing values
        df[col] = df[col].fillna(df[col].mean())

# Fill missing values for categorical columns with mode
for col in categorical_cols:
    if df[col].isnull().any():  # Check if column has missing values
        df[col] = df[col].fillna(df[col].mode()[0])

# Verify if any missing values remain
print("\nMissing Values After Imputation:")
print(df.isnull().sum())
```

```python
#4. Check for any duplicate values present in Dataset and do necessary to manage the duplicate items.{Hint: dataset.duplicated.sum()}
# Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Drop duplicates if any
if duplicate_count > 0:
    df = df.drop_duplicates()
    print(f"Duplicates removed. Remaining rows: {df.shape[0]}")
else:
    print("No duplicate rows found.")
```

```python
#Problem - 1:

import pandas as pd

# Loading the Titanic dataset
df = pd.read_csv('/content/drive/MyDrive/Concepts and Technology of AI/Titanic-Dataset.csv')

# Displaying the first few rows of the dataset
print(df.head())
```

```python
#Problem - 2:

# Subset for specified columns and first-class passengers
subset_df = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
first_class_df = subset_df[subset_df['Pclass'] == 1]

# Displaying the first few rows of the first-class subset
print(first_class_df.head())
```

```python
#Problem - 2:

mean_fare = first_class_df['Fare'].mean()
median_fare = first_class_df['Fare'].median()
max_fare = first_class_df['Fare'].max()
min_fare = first_class_df['Fare'].min()

print(f"Mean Fare: {mean_fare}")
print(f"Median Fare: {median_fare}")
print(f"Max Fare: {max_fare}")
print(f"Min Fare: {min_fare}")
```

```python
#Problem - 2:

null_age_count = first_class_df['Age'].isnull().sum()
print(f"Null values in Age column: {null_age_count}")

cleaned_df = first_class_df.dropna(subset=['Age'])
print(cleaned_df.head())
```

```python
#Problem - 3:
# Assuming original df has been loaded
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')

# Add new columns to original DataFrame
df = pd.concat([df, embarked_dummies], axis=1)

# Drop the original 'Embarked' column
df.drop('Embarked', axis=1, inplace=True)

# Display the modified DataFrame
print(df.head())
```

```python
#Problem - 4:

survival_rates_by_sex = df.groupby('Sex')['Survived'].mean()

# Visualization
import matplotlib.pyplot as plt

survival_rates_by_sex.plot(kind='bar', title='Mean Survival Rates by Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Problem - 5:

# Calculating mean survival rates by Embarked port
survival_rates_by_embarked = df.groupby(['Embarked_C', 'Embarked_Q', 'Embarked_S'])['Survived'].mean()

# Visualization
survival_rates_by_embarked.plot(kind='bar', title='Mean Survival Rates by Port of Embarkation')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Problem - 6{Optional}:

df['AgeGroup'] = pd.qcut(df['Age'], 5)

survival_rates_by_age_pclass = df.groupby(['AgeGroup', 'Pclass'])['Survived'].mean().unstack()

survival_rates_by_age_pclass.plot(kind='bar', title='Survival Rates by Age Group and Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
